## **1-3. ë°˜ë³µë¬¸ê³¼ ë¦¬ìŠ¤íŠ¸ (10ì )**

### **1-3-b. ë¦¬ìŠ¤íŠ¸ "alist" ì¤‘ ë‹¤ìŒ ì¡°ê±´ì— ë§ëŠ” ìš”ì†Œë§Œ ì¶”ì¶œí•˜ì—¬, ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ "blist"ë¥¼ ìƒì„±í•´ ì¶œë ¥í•´ ì£¼ì„¸ìš”. (3ì )**
- alist ì¤‘ ì•½ìˆ˜ê°€ (1)1, (2)ìì‹ ì˜ ì œê³±ê·¼, (3)ìê¸° ìì‹ ë§Œ ìˆëŠ” ìˆ«ìë§Œ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ "blist"ì— ë„£ì–´ ì£¼ì„¸ìš”.
- ì§ì ‘ ìˆ«ìë¥¼ ì°¾ì•„ ì…ë ¥í•˜ë©´ 0ì  ì²˜ë¦¬ë©ë‹ˆë‹¤.

````python
# ì¡°ê±´ì— ë”°ë¥´ë©´, ì†Œìˆ˜ì˜ ì œê³±ìˆ˜ë¥¼ ì¶œë ¥í•´ì•¼ í•¨: 4, 9, 25, 59

# ì†Œìˆ˜ì¸ì§€ íŒë³„í•˜ëŠ” í•¨ìˆ˜
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1): # 2ë¶€í„° nì˜ ì œê³±ê·¼ê¹Œì§€ë§Œ ë‚˜ëˆ„ì–´ í™•ì¸
        if n % i == 0:
            return False
    return True

# alistì˜ ê° ìˆ«ìì— ëŒ€í•´ ê·¸ ìˆ«ìì˜ ì œê³±ê·¼ì´ ì†Œìˆ˜ì¸ì§€ í™•ì¸í•´ blist ìƒì„±
alist = [4, 9, 16, 25, 36, 49]
blist = [num for num in alist if is_prime(int(num**0.5))]
print(blist)
````

## **3-1. ì œì‹œë¬¸ì— ë”°ë¼ ë¬¸ì œë¥¼ í•´ê²°í•´ ì£¼ì„¸ìš”. (5ì )**
- `division_point`ëŠ” ìˆ˜ì§ì„  ìœ„ì˜ ë‘ì ì˜ ì¢Œí‘œì™€ ë¹„ìœ¨ì„ ì…ë ¥ë°›ì•˜ì„ ë•Œ ë‚´ë¶„ì ì˜ ìœ„ì¹˜ë¥¼ êµ¬í•´ì£¼ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
- `division_point`í•¨ìˆ˜ë¥¼ ì•„ë˜ ì •ì˜ì— ë”°ë¼ êµ¬í˜„í•´ ì£¼ì„¸ìš”.
- ëª¨ë“  ë³€ìˆ˜ëŠ” ì–‘ìˆ˜ë¼ëŠ” ì „ì œ ì•„ë˜ í•¨ìˆ˜ë¥¼ ì •ì˜í•´ì£¼ì„¸ìš”.
###**ì…ë ¥ê°’**
  - point1 : ìˆ˜ì§ì„  ìœ„ ì²«ë²ˆì§¸ ì ì˜ ì¢Œí‘œ
  - point2 : ìˆ˜ì§ì„  ìœ„ ë‘ë²ˆì§¸ ì ì˜ ì¢Œí‘œ
  - m, n : point1ê³¼ point2 ì‚¬ì´ë¥¼ m : n ë¹„ìœ¨ë¡œ ë‚´ë¶„

  ### **ë°˜í™˜ê°’**
  - answer : ë‚´ë¶„ì ì˜ ì¢Œí‘œ

````python
## division_point í•¨ìˆ˜ ì •ì˜
def division_point(point1, point2, m, n):
    # ë‚´ë¶„ì  ê³µì‹: (n * point1 + m * point2) / (m + n)
    answer = (n * point1 + m * point2) / (m + n)
    return answer
````

````python
## í•¨ìˆ˜ í˜¸ì¶œ ë° ì‹¤í–‰
# í•´ë‹¹ ë¶€ë¶„ì€ ìˆ˜ì •í•˜ì‹œë©´ ì•ˆë©ë‹ˆë‹¤.
# ì‹¤í–‰í•´ì£¼ì„¸ìš”!

point1, point2, m, n = map(int, input('ë„¤ ìˆ˜ë¥¼ ê³µë°±ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì…ë ¥í•˜ì„¸ìš”: ').split())
print('##ì •ë‹µ##')
print(division_point(point1, point2, m, n))
````

### **3-2-1. ì‹­ì§„ìˆ˜ë¥¼ ì´ì§„ìˆ˜ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë°˜ë³µë¬¸ì„ ì´ìš©í•´ êµ¬í˜„í•´ì£¼ì„¸ìš”. (5ì )**

  #### **ì…ë ¥ê°’**
  - num : ì…ë ¥í•˜ëŠ” ì–‘ì˜ ì •ìˆ˜

#### **ë°˜í™˜ê°’**
  - inputì´ ë³€í™˜ëœ ì´ì§„ìˆ˜, type ìƒê´€ì—†ìŒ


````python
def binary_iterative(num):
  # ì…ë ¥ê°’ì´ 0ì´ë©´ "0"ì„ ë°”ë¡œ ë°˜í™˜
  if num == 0:
    return "0"

  binary_str = ""

  while num > 0:
    remainder = num % 2
    # ë‚˜ë¨¸ì§€ë¥¼ ì´ì§„ìˆ˜ ë¬¸ìì—´ ì•ì— ì¶”ê°€
    binary_str = str(remainder) + binary_str
    # ë‹¤ìŒ ê³„ì‚°ì„ ìœ„í•´ ëª«ì„ numì— ì €ì¥
    num = num // 2
  return binary_str
````

````python
### Testing
# ì´ ì…€ì€ ì‚­ì œ ë° ìˆ˜ì •í•˜ì‹œë©´ ì•ˆë©ë‹ˆë‹¤. ì‹¤í–‰í•´ì£¼ì„¸ìš”!

print(binary_iterative(100))  # ì¶œë ¥: 1100100
print(binary_iterative(10))  # ì¶œë ¥: 1010
print(binary_iterative(2))     # ì¶œë ¥: 10
````

### **3-2-2. ìœ„ì˜ í•¨ìˆ˜(**`binary_iterative`**)ë¥¼ ì¬ê·€í•¨ìˆ˜(**`binary_recursive`**)ë¡œ êµ¬í˜„í•´ì£¼ì„¸ìš”.(5ì )**



````python
def binary_recursive(num):
  # numì´ 1 ì´í•˜ì¼ ë•Œ ì¬ê·€ ì¤‘ë‹¨
  if num <= 1:
    return str(num)
  # ì¬ê·€: ëª«ìœ¼ë¡œ ì¬ê·€ í˜¸ì¶œí•œ ê²°ê³¼ ë’¤ì— ë‚˜ë¨¸ì§€ë¥¼ ë¬¸ìì—´ë¡œ ë¶™ì„
  else:
    return binary_recursive(num // 2) + str(num % 2)
````

````python
### Testing
# ì´ ì…€ì€ ì‚­ì œ ë° ìˆ˜ì •í•˜ì‹œë©´ ì•ˆë©ë‹ˆë‹¤. ì‹¤í–‰í•´ì£¼ì„¸ìš”!

print(binary_recursive(100))  # ì¶œë ¥: 1100100
print(binary_recursive(10))  # ì¶œë ¥: 1010
print(binary_recursive(2))     # ì¶œë ¥: 10
````

  - `profit()` : ìˆœìˆ˜ìµì„ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
  - ì•„ë˜ ì˜ˆì‹œì²˜ëŸ¼ ì¶œë ¥ë˜ê²Œ í•¨ìˆ˜ë¥¼ ì •ì˜í•´ì£¼ì„¸ìš”.

- ì¶œë ¥

ì´ë²ˆ ë‹¬ ì‚¬ë¬´ì‹¤ ë¹„í’ˆë¹„(ìœ¼)ë¡œ $100 ì§€ì¶œ, ì´ë²ˆ ë‹¬ ì´ ì§€ì¶œì€ $20100ì…ë‹ˆë‹¤.

````python
loss = 0 # ì´ë²ˆ ë‹¬ ì´ ì§€ì¶œ
gain = 0 # ì´ë²ˆ ë‹¬ ì´ ìˆ˜ì…

def expenditure(content, dollar):
  # global: í•¨ìˆ˜ ì™¸ë¶€ì— ìˆëŠ” loss ë³€ìˆ˜ ì‚¬ìš©ì„ ì„ ì–¸
  global loss
  loss += dollar
  print(f"{content}(ìœ¼)ë¡œ ${dollar} ì§€ì¶œ, ì´ë²ˆ ë‹¬ ì´ ì§€ì¶œì€ ${loss}ì…ë‹ˆë‹¤.")

def income(content, dollar):
  # global: gain ë³€ìˆ˜ ì‚¬ìš©ì„ ì„ ì–¸
  global gain
  gain += dollar
  print(f"{content}(ìœ¼)ë¡œ ${dollar} ì…ê¸ˆ, ì´ë²ˆ ë‹¬ ì´ ìˆ˜ì…ì€ ${gain}ì…ë‹ˆë‹¤.")

def profit():
  # ìˆœì´ìµ = ì´ ìˆ˜ì…-ì´ ì§€ì¶œ
  net_profit = gain - loss
  print(f"ì´ë²ˆ ë‹¬ ìˆœì´ìµì€ ${net_profit}ì…ë‹ˆë‹¤.")
````
